# Bonus (solved with the help of many tools : stackoverflow, bingchat, chatgpt)

import pandas as pd

def clean_data(file_path):
    # load the dataset into a DataFrame
    file1 = pd.read_csv(file_path)
    
    # convert column names to lowercase and replace characters
    file1.columns = file1.columns.str.lower().str.replace('0', '').str.replace(':', '').str.replace(' ', '_')
    
    # drop the 'unnamed' column and rows with 'unemployed' employment status
    file1.drop(columns='unnamed', inplace=True)
    file1 = file1[file1['employmentstatus'] != 'Unemployed']
    
    # fill missing values in 'vehicle_type' column with 'U'
    file1['vehicle_type'].fillna('U', inplace=True)
    
    # drop rows with missing values and convert 'effective_to_date' column to datetime
    file1.dropna(inplace=True)
    file1['effective_to_date'] = pd.to_datetime(file1['effective_to_date'], errors='coerce')
    
    # extract the months from 'effective_to_date' column and filter for first quarter
    file1 = file1.assign(month=file1['effective_to_date'].dt.month)
    file1 = file1[(file1['month'] >= 1) & (file1['month'] <= 3)]
    
    return file1
